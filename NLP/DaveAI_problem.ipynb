```python
from itertools import product
from string import ascii_lowercase
import time

"""
keywords = [''.join(i) for i in product(ascii_lowercase, repeat = 3)]
combinations=list(product(numbers,repeat=2))

"""
"""
Code uses cross product of all possible indices combinations for (insertion,deletion and substitution ) with all
possible combinations of alphabetical (replacement and substitution sets) . Merging these cross-products with 
the base-word gives the desired result.
<itertools.product> method is used for mapping .
<time.time()> is used to get an assessment of run-time.

"""

def combination(size,n):
    if n==0:
        print('0 combinations are not possible! Try a number greater than 0 !')
    combinations=list(product(list(range(size)),repeat=n))
    return combinations

    

def keywords(n):
    if n==0:
        print('0 combinations are not possible! Try a number greater than 0 !')
    #combos_str=[''.join(i) for i in product(ascii_lowercase, repeat = n)]
    combos_list=product(ascii_lowercase, repeat = n)
    return combos_list
        
#add,sub,del -mistakes.
        
def sub(word,n):
    if n==0:
        return []
    
    comb=combination(len(word),n)
    
    #to have all n substitutions to be unique : (1,1) is 2-sub but happening on same element which is a mistake
    sub_combos=[]
    sub_combos.extend(comb)
    for combo in comb:
        contains_duplicates = any(combo.count(element) > 1 for element in combo)
        if contains_duplicates:
            sub_combos.remove(combo)
            
    result=[]    
    for x in sub_combos:
        replacements=keywords(n)
        for y in replacements:
            a=list(zip(x,y))
            temp=word
            for index,replacement in a:
                temp=temp[:index]+replacement+temp[index+1:]
            result.append(temp)

    result=[a for a in result if a!=word]
    return result
            
    

def insert(word,n):
    if n==0:
        return []
    ins_combos=combination(len(word)+1,n)
    result=[]
    for x in ins_combos:
        replacements=keywords(n)
        for y in replacements:
            a=list(zip(x,y))
            temp=word
            for index,replacement in a:
                temp=temp[:index]+replacement+temp[index:]
            result.append(temp)

    result=[a for a in result if a!=word]
    return result
    
    
def delete(word,n):
    if n==0:
        return []
    result=[]
    length=len(word)
    combos=combination(length,n)
    
    
    #(1,1,0) is actually 1 deletion instead of 2. so these combinations must be removed for perfection.
    indices=[]
    indices.extend(combos)
    for combo in combos:
        contains_duplicates = any(combo.count(element) > 1 for element in combo)
        if contains_duplicates:
            indices.remove(combo)
    
           
    for slice in indices:  
        codes=[]
        for index in slice:
            codes.append(f' i != {index} ')
            codes.append('and')
        if codes[-1]=='and':
            codes=codes[:-1]
        code=''.join(codes)
        temp =''.join([word[i] for i in range(len(word)) if eval(code)]) 
        result.append(temp)
    return list(set(result))  # to remove duplicates resulting from indices,say (1,0,2) (2,0,1) (0,2,1) 
                              # because all these slices will remove same characters only.
    
def functional_combos(*a):
    b=max(a)
    c=list(product(list(range(0,b+1)),repeat=3))

    if a[0]==0:
        c=[(0,j,k) for (_,j,k) in c]
        c=list(set(c))
    if a[1]==0:
        c=[(i,0,k) for (i,_,k) in c]
        c=list(set(c))
    if a[2]==0:
        c=[(i,j,0) for (i,j,_) in c]
        c=list(set(c))
    c=[(i,j,k) for (i,j,k) in c if i<=a[0] and j<=a[1] and k<=a[2]]
    c=[item for item in c if item!=(0,0,0)]
    return c
        

def variate(word, number_of_insertions = 2, number_of_deletions = 2, number_of_substitutions = 2):
    
    function_pairs=functional_combos(number_of_insertions , number_of_deletions, number_of_substitutions )
    
    for a,b,c in function_pairs:
        print(f'{a}-insertion\n{b}-deletion\n{c}-substitution : ')
        print('\n')
        result1=insert(word,a)
        result2=delete(word,b)
        result3=sub(word,c)
        result=[]
        result.extend(result1)
        result.extend(result2)
        result.extend(result3)
        print(result)
        print('\n')
    return

start=time.time()
variate('and',2,2,2)
print(f'Total time taken is : {time.time()-start} s')


```
